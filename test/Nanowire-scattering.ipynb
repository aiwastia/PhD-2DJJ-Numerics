```python
import numpy as np
import cmath
from scipy import optimize as opt
from matplotlib import pyplot as pp
```

```python
#En=0, kx=0

m=0.5
muSC=15.
W=10**-7
Wbis=3.

def fdetB(B):
    muSC=mu
    lam=B*W/cmath.sqrt(2*muSC/m/10.)
    al=-1j
    ke=cmath.sqrt(2*m*(mu+B))
    Kh=cmath.sqrt(2*m*(B-mu))
    #kSC=cmath.sqrt(2*m*muSC)
    #kpe=cmath.sqrt(2*m*(muSC+B))
    #kph=cmath.sqrt(2*m*(muSC-B))
    #NNe=cmath.cos(kpe*W)-1j*cmath.sin(kpe*W)*(kSC**2+kpe**2)/(2*kSC*kpe)
    #NNh=cmath.cos(kph*W)+1j*cmath.sin(kph*W)*(kSC**2+kph**2)/(2*kSC*kph)
    re=1./(1j/lam-1)
    #re=-1j*(kSC**2-kpe**2)/(2*kSC*kpe)*cmath.sin(kpe*W)/NNe
    te=1./(1j*lam+1)
    #te=cmath.exp(-1j*kSC*W)/NNe
    rh=1./(-1j/lam-1)
    #rh=1j*(kSC**2-kph**2)/(2*kSC*kph)*cmath.sin(kph*W)/NNh
    th=1./(-1j*lam+1)
    #th=cmath.exp(1j*kSC*W)/NNh
    A=re**2-te**2
    AA=rh**2-th**2
    wwv=1+re*cmath.exp(2*1j*ke*Wbis)+rh*np.exp(2*Kh*Wbis)-al**2*(re*rh+np.exp(2*1j*ke*Wbis)*rh*A+np.exp(2*Kh*Wbis)*re*AA+\
                                                           np.exp(2*1j*ke*Wbis+2*Kh*Wbis)*A*AA)
    #print(wwv)
    return np.absolute(wwv)

mumin=0.
mumax=10.
muarray=np.linspace(mumin,mumax,np.int((mumax-mumin)/0.1)+1)

BmuList=[0]*len(muarray)
for j in range(len(muarray)):
    mu=muarray[j]
    print('{}) mu={}'.format(j+1,mu))
    Bmin=0.
    Bmax=2.*mu
    x0roots=np.linspace(Bmin,Bmax,np.int((Bmax-Bmin)/.1))
    Barray=[]
    for i in range(len(x0roots)):
        print('{}-{}'.format(j+1,i+1))
        x0=x0roots[i]
        solekx=opt.fsolve(fdetB,np.asarray(x0))
        if abs(fdetB(solekx))>10**-10 or solekx<=0:
            solekx=[]
        solekx=np.ndarray.tolist(np.around(solekx,6))
        print(solekx)
        Barray.extend(solekx)
    Barray=sorted(list(set(Barray)))
    BmuList[j]=Barray
```

```python
SelInd=np.where(BmuList)[0]
mulist2=[muarray[i] for i in SelInd]
BmuList2=[BmuList[i][0] for i in SelInd]
```

```python
pp.figure(figsize=(20,15))
for i in range(len(BmuList)):
    pp.plot([muarray[i]]*len(BmuList[i]),BmuList[i],'.')

pp.plot(muarray, muarray)
#M,T=7,1
#pp.plot(muarray, [np.sqrt(pop**2+(M*T*np.pi**2/8/m/Wbis**2)**2) for pop in muarray])
#pp.axis([0,9,1.25,1.6])
pp.title('det=0 (4x4) for scattering in a nanowire')
pp.xlabel('$\mu$',fontsize=15)
pp.ylabel('$B$',fontsize=15)
```

```python
import numpy as np
from matplotlib import pyplot as pp
import cmath
```

```python
def cot(x):
    return 1/np.tan(x)
```

```python
m=0.5
mu=1.8
t=100.
a=1/np.sqrt(2*m*t)
W=21*a #np.pi
tp=1.

PiNu0=m/2
PiNu=1/(2*m*a)
gam=PiNu0/PiNu*tp**2
```

```python
#functions at zero energy
#expansion in 'a' which is the distance to the boundary

def kBe(B):return cmath.sqrt(2*m*(mu-B))
def kBh(B):return np.sqrt(2*m*(mu+B))

def g11e(B):return -(1-a*kBe(B)*cot(kBe(B)*W)-2./3*(a*kBe(B))**2)
def g11h(B):return 1-a*kBh(B)*cot(kBh(B)*W)-2./3*(a*kBh(B))**2
def g1Ne(B):return -(a*kBe(B)/np.sin(kBe(B)*W))
def g1Nh(B):return a*kBh(B)/np.sin(kBh(B)*W)

def alpha(B):return g11e(B)**2*g11h(B)**2+g1Ne(B)**2*g1Nh(B)**2-g11h(B)**2*g1Ne(B)**2-g1Nh(B)**2*g11e(B)**2

def COSphi(B):return ((alpha(B)*gam**4+1)/2/gam**2-g11e(B)*g11h(B))/g1Ne(B)/g1Nh(B)
```

```python
Barray=np.linspace(0,30,50001)
#phiarray=[np.arccos(COSphi(B)) for B in Barray]
```

```python
def detBPHI(B,phi):return alpha(B)*gam**4-2*gam**2*(g11e(B)*g11h(B)+g1Ne(B)*g1Nh(B)*np.cos(phi))+1
```

```python
phi=np.pi*1.75
detarray=[detBPHI(B,phi) for B in Barray]
```

```python
pp.plot(Barray,detarray)
pp.axis([0,max(Barray),-5,5])
```
